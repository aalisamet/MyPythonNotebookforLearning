# Practical Functions

````python
def divideNumbers(number):
    return number/2
myList = [3,7,5,10,20,30]
for i in mylist:
    print(devideNumbers(i))
````

````python
# Map functions aslinda amaci bir listeyi bir kurala gore baska bir listeye donusturmek icin kullanilir

newMyList = list(map(divideNumbers,myList))
print(newMyList)
````

````python
def controlString(controlStringArg):
    return "Atil" in controlStringArg

if (controlString("denemeAtild")):
    print("Geciyor")
else:
    print("Gecmiyor")
````

````python
# filter aslinda java steamdeki filter ile ayni mantikta calisir bir pipeline olusturur ve 
# Liste burada akarken filter sonucu false cikanlari pipelinedan temizler

filter(controlString,newMyList) # gibi bir kullanimi vardir
````
